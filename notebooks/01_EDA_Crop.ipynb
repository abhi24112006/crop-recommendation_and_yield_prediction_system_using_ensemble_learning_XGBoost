```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/Crop_recommendation.csv")

df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="label", data=df)
plt.xticks(rotation=90)
plt.title("Crop Distribution")
plt.show()
```

Dataset is balanced.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()

```

Correlation matrix is computed only for numerical variables. Since the target variable contains categorical crop labels, we exclude non-numeric columns before computing Pearson correlation.

1. Rainfall vs humidity correlation.
2. NPK interelation.

```python
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```
