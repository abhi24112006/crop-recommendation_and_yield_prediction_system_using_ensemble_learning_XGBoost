```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ["Logistic", "Decision Tree", "Random Forest", "XGBoost"]

accuracy = [0.96, 0.97, 0.993, 0.994]
precision = [0.96, 0.97, 0.993, 0.994]
recall = [0.96, 0.97, 0.993, 0.994]
f1_score = [0.96, 0.97, 0.993, 0.994]
```

```python
plt.figure()
plt.bar(models, accuracy)
plt.title("Accuracy vs Models")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.figure()
plt.bar(models, precision)
plt.title("Precision vs Models")
plt.xlabel("Models")
plt.ylabel("Precision")
plt.show()
```

```python
plt.figure()
plt.bar(models, precision)
plt.title("Precision vs Models")
plt.xlabel("Models")
plt.ylabel("Precision")
plt.show()
```

```python
plt.figure()
plt.bar(models, recall)
plt.title("Recall vs Models")
plt.xlabel("Models")
plt.ylabel("Recall")
plt.show()
```

```python
x = np.arange(len(models))
width = 0.2

plt.figure()

plt.bar(x - 1.5*width, accuracy, width, label='Accuracy')
plt.bar(x - 0.5*width, precision, width, label='Precision')
plt.bar(x + 0.5*width, recall, width, label='Recall')
plt.bar(x + 1.5*width, f1_score, width, label='F1 Score')

plt.xticks(x, models)
plt.title("Model Comparison")
plt.xlabel("Models")
plt.ylabel("Score")
plt.legend()

plt.show()
```

```python
plt.figure()
plt.bar(models, f1_score)
plt.title("F1 Score vs Models")
plt.xlabel("Models")
plt.ylabel("F1 Score")
plt.show()
```
